```python
#Let's create pseudo absence points using the presence data
#Reference: 
#1) https://www.rsgislib.org/rsgislib_sdm.html
#2) https://github.com/dydwkd486/SDMsBaggingEnsemble/blob/main/preprocessing/mopa_SDM.r
#3) https://rspatial.org/raster/sdm/3_sdm_absence-background.html
#4) https://www.researchgate.net/post/Do_you_know_any_Python_library_for_Species_Distribution_Modeling


import rsgislib.sdm
import random
random.seed(123)

# Define simulation parameters within a Python dictionary:
# These are some of my bioclimatic variables
ContinuousLayers = ['data/bio_current/bio1_clipped.tif',
                    'data/bio_current/bio3_clipped.tif',
                   'data/bio_current/bio7_clipped.tif',
                   'data/bio_current/bio8_clipped.tif',
                   'data/bio_current/bio9_clipped.tif',
                   'data/bio_current/bio12_clipped.tif',
                   'data/bio_current/bio13_clipped.tif',
                   'data/bio_current/bio14_clipped.tif',
                   'data/bio_current/bio15_clipped.tif']


simulation = {}
simulation['Name'] = 'Absences'
simulation['Output Directory'] = './' + simulation['Name']
simulation['Images'] = ContinuousLayers 
simulation['Null Value'] = 0  # No data value for the Continuous and Categorical raster layers.
simulation['Valid Mask'] = 'data/sing_map.tif'

# define the study region using the binary mask image:
geotransform, projection, xsize, ysize, noDataVal = rsgislib.sdm.read_raster_information(simulation['Valid Mask'])
simulation['geotransform'] = geotransform
simulation['projection'] = projection
simulation['rasterxsize'] = xsize
simulation['rasterysize'] = ysize
del geotransform, projection, xsize, ysize, noDataVal
```

```python
# read xy coordinates for the presence records:
presence_xy = rsgislib.sdm.read_sightings_csv('data/csv/mousedeer_point.csv', sep=',', xfield='longitude', yfield='latitude')  # import from CSV file.

# extract values from input images for presence records:
presence_data = rsgislib.sdm.extract_raster_values(simulation, presence_xy)

# generate pseudo-absence x&y coordinates:
absence_xy = rsgislib.sdm.generate_pseudoabsences(simulation, n_points=3000)

# ensure that number of presence records == number of absence records:
absence_xy = rsgislib.sdm.equalise_records(presence_xy, absence_xy)
```

```python
#Let's create a dataframe containing all the absence data
absence = pd.DataFrame(absence_xy,
                      columns =['longitude','latitude'])
```

```python
#Label it with 0
absence["presence"] = 0
```

```python
### convert dataframe into geodataframe -- here we are creating POINT geodata
absence_point= gpd.GeoDataFrame(absence, 
                                       geometry=gpd.points_from_xy(absence.longitude, absence.latitude))
```

```python
#Visualize presence and absence data
fig, ax = plt.subplots(figsize=(16,38))
singapore_map.plot(ax=ax, color = "grey")
mousedeer_point.plot(categorical=True,
           legend=True,
           figsize=(10,6),
           markersize=40,
           marker='o',
           c ="blue",
           ax=ax)
absence_point.plot(categorical=True,
           legend=True,
           figsize=(10,6),
           markersize=40,
           marker='o',
           c ="red",
           ax=ax)
ax.set_title('Mousedeer presence in SG')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.axis()
plt.show()
```

```python
#Combine presence and absence data into one dataframe
presence_absence = pd.concat([mousedeer_point,absence_point],ignore_index=True)
```

## Modelling and Evaluation
### Current bioclimatic variables

```python
#Let's set up the parameters for our model
#Note there is a difference between _clipped and _clip
#The raster files needed to have no data values '-999', imputed in to be
#read for training data

#This is using current bioclimatic variables
X_rasters = ['data/bio_current/bio1_clip.tif',
             'data/bio_current/bio2_clip.tif',
             'data/bio_current/bio3_clip.tif',
             'data/bio_current/bio4_clip.tif',
             'data/bio_current/bio5_clip.tif',
             'data/bio_current/bio6_clip.tif',
             'data/bio_current/bio7_clip.tif',
             'data/bio_current/bio8_clip.tif',
             'data/bio_current/bio9_clip.tif',
             'data/bio_current/bio10_clip.tif',
             'data/bio_current/bio11_clip.tif',
             'data/bio_current/bio12_clip.tif',
             'data/bio_current/bio13_clip.tif',
             'data/bio_current/bio14_clip.tif',
             'data/bio_current/bio15_clip.tif',
             'data/bio_current/bio16_clip.tif',
             'data/bio_current/bio17_clip.tif',
             'data/bio_current/bio18_clip.tif',
             'data/bio_current/bio19_clip.tif']

Y_vector = gpd.read_file('pa.shp')
```

```python
#Reference: 
#1) https://github.com/perrygeo/pyimpute
#2) https://daniel-furman.github.io/Python-species-distribution-modeling/
#3) https://www.youtube.com/watch?v=7mmbRNE9VsA

#train-test-split
train_xs,train_y = load_training_vector(Y_vector,X_rasters,
                                        response_field='presence')
```

```python
# Check shape
target_xs, raster_info = load_targets(X_rasters)
train_xs.shape, train_y.shape 
```

```python
#Initializing model for RandomTree Classifier
clf = RandomForestClassifier()
clf.fit(train_xs,train_y)
```

```python
#Initializing model for KNN Classifier
knn = KNeighborsClassifier()
knn.fit(train_xs,train_y)
```

```python
#Initializing model for SVM
svm = SVC()
svm.fit(train_xs,train_y)
```

```python
#Initializing model for LogisticRegression
lr = LogisticRegression()
lr.fit(train_xs,train_y)
```

```python
# Evaluate model RandomForest
evaluate_clf(clf,train_xs,train_y)
```

```python
# Evaluate model KNN
evaluate_clf(knn,train_xs,train_y)
```

```python
# Evaluate model SVM
evaluate_clf(svm,train_xs,train_y)
```

```python
# Evaluate model Logistis regression
evaluate_clf(lr,train_xs,train_y)
```
